```python
# Loading the dataset
train=pd.read_csv('/content/drive/MyDrive/data/Automobile_data (1).csv')
```

```python
#  Display the first few rows of the dataset to get an idea of its structure
train.head()
```

```python
# To check mean of data set......The describe() method provides a detailed statistical
# summary of dataset, which includes the mean among other statistics like count,
# standard deviation (std), minimum (min), and maximum (max) values, as well as the quartiles (25%, 50%/median, 75%).
train.describe()
```

```python
# Show information of dataset
train.info()
```

```python
# To check shape of dataset
train.shape
```

```python
# To check duplicates
train.duplicated()
```

```python
# To remove duplicates
train.drop_duplicates()
```

```python
# To check shape again
train.shape
```

```python
# Replace "?" with NaN
train.replace('?', np.nan)
```

```python
train.isnull().sum()
```

```python
train.shape
```

```python
#imputation
train['normalized-losses'] = pd.to_numeric(train['normalized-losses'], errors='coerce')
train['normalized-losses'].mean()
train['normalized-losses'] = train['normalized-losses'].fillna(train['normalized-losses'].mean())
```

```python
# Check for missing values
train.isnull().sum()
```

```python
# Drop Rows with Missing Values in the "price" Column:
auto_cleaned = train.dropna(subset=['price'])
```

```python
# Check the Shape of data After Dropping Rows
auto_cleaned.shape
```

```python
# Display Column Names After Dropping Rows
auto_cleaned.describe()
```

```python
auto_cleaned.columns
```

```python
# Finding outliers in the data by plotting (normalized losses and symbolling)
sns.boxplot(x='symboling', y='normalized-losses',data=auto_cleaned)
```

```python
# Checking distribution of data using distribution plot.
sns.distplot(auto_cleaned['normalized-losses'])
```

```python
# Droping outliers
titanic_cleaned= auto_cleaned[auto_cleaned['normalized-losses']<150]
```

```python
# Verifying that outliers removed or not
sns.boxplot(y='normalized-losses',data=titanic_cleaned)
```
